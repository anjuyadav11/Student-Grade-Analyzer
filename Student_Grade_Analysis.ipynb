# **Student Grade Analysis**

-------------

## **Objective**

To develop a Python program that analyzes student grades and provides insights.


## **Import Library**

```python
import pandas as pd

import matplotlib.pyplot as plt
```

## **Import Data**

```python
data = pd.read_csv (r"C:\Users\acer\Downloads\students_data.csv") 

#display the data
print(data)
```

## **Describe Data**

```python
df = pd.DataFrame(data)
```

## **Sort the Data**

```python
df_sorted = df.sort_values(by='Average', ascending=False)
```

## **Data Preprocessing**

```python
print("Top 3 Students:")

print(df_sorted.head(3))
```

## **Data Visualisation**

```python
plt.bar(df_sorted['Roll Number'], df_sorted['Average'])

plt.xlabel('Roll Number')

plt.ylabel('Average Grade')

plt.title('Student Grade Analysis')

plt.show()
```

## **Explaination**

This project demonstrates basic data analysis and visualization using Python, it imports student grades from csv files sort them and list the required data. We can also visually represent the data with the help of graphs.
